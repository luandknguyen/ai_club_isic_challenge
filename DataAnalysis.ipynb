```python
import os
import csv

import numpy
from PIL import Image
from matplotlib import pyplot
```

```python
IMAGE = "../datasets/training/images/ISIC_0000000.jpg"
LABEL = "../datasets/training/segmentation/ISIC_0000000_segmentation.png"
INDEX = "index.csv"
WIDTH = 1024
HEIGHT = 1024
SIZE = (WIDTH, HEIGHT)
```

```python
image = Image.open(IMAGE)
pyplot.imshow(image)
image = image.resize(SIZE)
```

```python
label = Image.open(LABEL)
pyplot.imshow(label)
label = label.resize(SIZE)
```

```python
training_data = numpy.array(image.getdata(), dtype="float").reshape(1, WIDTH, HEIGHT, 3)
label_data = numpy.array(label.getdata(), dtype="float").reshape(1, WIDTH, HEIGHT, 1)
print(f"Training data shape: {training_data.shape}")
print(f"Label data shape: {label_data.shape}")
```
